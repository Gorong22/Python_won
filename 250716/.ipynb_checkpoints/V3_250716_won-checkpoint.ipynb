```python
# 미션
# 공공데이터포털 사이트 -> 캠핑장 정보를 api로 찾아옴 
# 100개의 캠핑장 정보만 엑셀에 저장한다 
# 필수데이터가 있음 : 캠핑장 아이디 (num, contentID, 캠핑장이름(facltNm), 위치(doNm), 주소(address1),홈페이지주소(homepage))
# 엑셀에 저장
```

```python
import requests
import openpyxl

deco = "vXpamPPqXWsORdlj04YF6P/II0iMXCy2cV5WrUe78yyyLIH3JSQ5Fl82BQb0YzAJwwhceO+xn/ae1OvcA7pXmg=="

camping_open_api = "http://apis.data.go.kr/B551011/GoCamping/basedList"
params = {
    "serviceKey": deco,
    "numOfRows": "100",
    "pageNo": "1",
    "MobileOS": "ETC",    
    "MobileApp": "camnavi",
    "_type": "json"
}
num= 1

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions["B"].width=100
excel_sheet.column_dimensions["C"].width=100
excel_sheet.column_dimensions["D"].width=100
excel_sheet.column_dimensions["E"].width=100
excel_sheet.column_dimensions["F"].width=100
excel_sheet.column_dimensions["G"].width=100
excel_sheet.append(["No", "ContentId", "캠핑장명", "주소", "링크"])



res = requests.get(camping_open_api, params = params)

if res.status_code == 200:
    data = res.json()
    camp_list = data["response"]["body"]["items"]["item"]

    for camp in camp_list :
        num += 1
        excel_sheet.append([num, camp["contentId"], camp["facltNm"], camp["doNm"], camp["addr1"], camp["homepage"]])
        print(num, camp["contentId"], camp["facltNm"], camp["doNm"], camp["addr1"], camp["homepage"])
else:
    print("에러발생", res.status_code)
excel_file.save("campingList_with_dataPotal.xlsx")
excel_file.close()
```
