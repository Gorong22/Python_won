```python
#삼성전자 뉴스룸 최신기사 크롤링

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument('--window-size=1920x1080')
options.add_argument('--start-maximized')
options.add_argument('--Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Mobile Safari/537.36'
)
options.add_argument('--lang=ko_KR')

driver = webdriver.Chrome(service=service, options=options)

url = "https://news.samsung.com/kr/"
driver.get(url)

samsung = driver.find_elements(By.CSS_SELECTOR, "ul#recent_list_box > li")

for i, samsungs in enumerate(samsung, start =1):
    title = samsungs.find_element(By.CSS_SELECTOR, "span.ellipsis")
    print(f"{i}-{title.text}")
```

```python
#롯데시네마에서 현재 상영작 제목을 수집하세요

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument('--window-size=1920x1080')
options.add_argument('--start-maximized')
options.add_argument('--Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Mobile Safari/537.36'
)
options.add_argument('--lang=ko_KR')

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.lottecinema.co.kr/NLCHS/Movie/List?flag=1"
driver.get(url)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)

lotte = driver.find_elements(By.CSS_SELECTOR, "ul.movie_list.type2 > li.screen_add_box")
time.sleep(2)

for i, movies in enumerate(lotte, start=1) :
    title = movies.find_element(By.XPATH, ".//div[contains(@class, 'btm_info')][.//strong[contains(@class,'tit_info')]]")
    print(f"{i}-{title.text}")
```

```python
#GPT 형님 답변
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# 크롬 옵션 설정
options = Options()
options.add_argument("--window-size=1920,1080")
options.add_argument("--lang=ko_KR")

# 드라이버 실행
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 롯데시네마 상영작 페이지 이동
url = "https://www.lottecinema.co.kr/NLCHS/Movie/List?flag=1"
driver.get(url)

time.sleep(2)  # 페이지 로딩 대기

# 영화 아이템(li) 전체 가져오기
lotte = driver.find_elements(By.CSS_SELECTOR, "ul.movie_list.type2 > li.screen_add_box")

print(f"총 li 개수: {len(lotte)}\n")

# li별로 제목 추출
for i, movies in enumerate(lotte, start=1):
    title_el = movies.find_elements(
        By.XPATH,
        ".//div[contains(@class, 'btm_info')]/strong[contains(@class,'tit_info')]"
    )
    if title_el:  # 제목이 있는 경우만 출력
        title = title_el[0].get_attribute("innerText").strip()
        print(f"{i} - {title}")

driver.quit()
```

```python
#노컷뉴스 연예면 주요 기사 제목 5개 수집하기 
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
options = Options()
options.add_argument('--window-size=1920x1080')
options.add_argument('--start-maximized')
options.add_argument('--Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Mobile Safari/537.36'
)
options.add_argument('--lang=ko_KR')

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.nocutnews.co.kr/news/entertainment"
driver.get(url)
time.sleep(3)

title_a = driver.find_elements(By.CSS_SELECTOR, "li.li")

for i, title in enumerate(list(title_a)[:5], start=1):
    title_1 = title.find_element(By.XPATH, ".//a[contains(@class, 'img_n')]/p[contains(@class, 'txt')]/strong[contains(@class, 'clamp2')]")
    print(f"{i}-{title_1.text}")
```

```python
#아프라카 tv 제목만 크롤링하기 
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time 

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument('--window-size=1920x1080')
options.add_argument('start-maximized')
options.add_argument('--Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Mobile Safari/537.36'
)
options.add_argument('--lang=ko_KR')

driver = webdriver.Chrome(service=service, options=options)

url = "https://www.sooplive.co.kr/"
driver.get(url)
time.sleep(5)

soop = driver.find_elements(By.XPATH, ".//div[@class='cBox-list']//li[contains(@data-type, 'cBox')]")


for soops in soop :
    title = soops.find_element(By.XPATH, ".//div[contains(@class, 'cBox-info')]//h3[@class='title']/a")
    print(title.text)
```
