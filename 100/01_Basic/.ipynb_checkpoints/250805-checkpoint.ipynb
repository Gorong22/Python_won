```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
#from webdriver_manager.common.keys import keys
from webdriver_manager.chrome import ChromeDriverManager
import time

service = Service(ChromeDriverManager().install())
options = Options()

options.add_argument('--lang=ko_KR')
driver = webdriver.Chrome(service=service, options=options)

driver.get("https://brunch.co.kr/")
time.sleep(3)

for page_num in range(9) :
    try :
        page_button = driver.find_element(By.CSS_SELECTOR, "div.wrap_btn  button.btn_scroll.btn_brunch.btn_next")
        page_button.click()
        time.sleep(2)
        titles = driver.find_elements(By.CSS_SELECTOR, ".append_info")
        for title in titles :
            main = title.find_element(By.CSS_SELECTOR, "strong.tit_pic")
            print(main.text.strip())
            
    except Exception as e:
        print(f"{page_num} 페이지 클릭중 오류:", e)
        break

        

driver.quit()

```
