```python
import requests
from bs4 import BeautifulSoup

url = "https://stock.mk.co.kr/domestic/all_stocks?type=kospi&status=industry"

res = (requests.get(url))
soup = BeautifulSoup(res.content, "html.parser")
title = soup.select("ul.table-stock.sty1>li.row_sty")

for titles in title :
    company = titles.select_one("div.st_name").get_text().strip()
    price = titles.select_one("div.st_price").get_text().strip()
    rate = titles.select_one("div.st_rate").get_text().strip()
    print(f"{company}, {price}, {rate}")
```
