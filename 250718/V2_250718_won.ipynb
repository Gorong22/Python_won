```python
import requests, openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font

wrap_alignment = Alignment(wrap_text=True)
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
normal_font = Font(bold=False)



excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = "게시글"


header = ["No", "게시글 제목", "댓글수"]
excel_sheet.append(header)

for cell in excel_sheet[1] :
    cell.font = header_font

excel_sheet.column_dimensions["A"].width = 20
excel_sheet.column_dimensions["B"].width = 80
excel_sheet.column_dimensions["C"].width = 80

url = "https://davelee-fun.github.io/trial/board/news.html"

res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
title2 = soup.select("div.list_content>.list_item")
num = 0
row = 2

for titles2 in title2 :
    title0 = titles2.select_one("span.subject_fixed")
    reply_count = titles2.select_one("span.rSymph05")
    link = titles2.select_one("a.list_subject")
#div_content > div.list_content > div:nth-child(1) > div.list_title > a.list_reply.reply_symph
    
    if titles2 != None :
        num += 1
        #print(f"{num}. {title0.get_text(strip=True)}{[int(reply_count.get_text(strip=True))]}")
        link_url = "https://davelee-fun.github.io/trial/board/"+link["href"]
        data1 = [num, title0.get_text(strip=True), "[" + reply_count.get_text(strip=True) + "]"]
        excel_sheet.append(data1)
        excel_sheet.cell(row = row, column = 2).hyperlink = link_url
        for cell in excel_sheet[row] :
            cell.font = bold_font
        row += 1                  
        res_title = requests.get("https://davelee-fun.github.io/trial/board/"+link["href"])
        soup_title = BeautifulSoup(res_title.content, "html.parser")
        replies = soup_title.select("div.comment_content")
        for reply in replies :
            data2 = ["", "", reply.get_text(strip=True).replace(" ","").replace("\n", "").replace("\t", "")]
            #print("  ㄴ", num, reply.get_text(strip=True).replace(" ","").replace("\n", "").replace("\t", ""))
            excel_sheet.append(data2)
        for cell in excel_sheet[row] :
            cell.font = normal_font
            cell.alignment = wrap_alignment
        row += 1
excel_file.save("mysite.xlsx")
excel_file.close()
print(f"mysite.xlsx 저장완료")
```

```python
import requests, openpyxl
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font, colors

wrap_alignment = Alignment(wrap_text=True)
center_alignment = Alignment(horizontal="center", vertical="center")
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
normal_font = Font(bold=False)
hyperlink_font = Font(color=colors.BLUE, underline="single")


excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = "게시글"


header = ["No", "게시글 제목", "댓글수"]
excel_sheet.append(header)

for cell in excel_sheet[1] :
    cell.font = header_font
    cell.alignment = center_alignment

excel_sheet.column_dimensions["A"].width = 20
excel_sheet.column_dimensions["B"].width = 80
excel_sheet.column_dimensions["C"].width = 80

url = "https://davelee-fun.github.io/trial/board/news.html"

res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")
title2 = soup.select("div.list_content>.list_item")
num = 0
row = 2

for titles2 in title2 :
    title0 = titles2.select_one("span.subject_fixed")
    reply_count = titles2.select_one("span.rSymph05")
    link = titles2.select_one("a.list_subject")
#div_content > div.list_content > div:nth-child(1) > div.list_title > a.list_reply.reply_symph
    
    if titles2 != None :
        num += 1
        #print(f"{num}. {title0.get_text(strip=True)}{[int(reply_count.get_text(strip=True))]}")
        link_url = "https://davelee-fun.github.io/trial/board/"+link["href"]
        data1 = [num, title0.get_text(strip=True), "[" + reply_count.get_text(strip=True) + "]"]
        excel_sheet.append(data1)
        excel_sheet.cell(row = row, column = 2).hyperlink = link_url
        excel_sheet.cell(row = row, column=2).font = hyperlink_font
        
        for cell in excel_sheet[row] :
            if cell.column != 2 :
                cell.alignment = wrap_alignment  
                cell.font = bold_font
     
        row += 1                  
        res_title = requests.get("https://davelee-fun.github.io/trial/board/"+link["href"])
        soup_title = BeautifulSoup(res_title.content, "html.parser")
        replies = soup_title.select("div.comment_content > div.comment_view")
        for reply in replies :
            data2 = ["", "", reply.get_text(strip=True).replace(" ","").replace("\n", "").replace("\t", "")]
            #print("  ㄴ", num, reply.get_text(strip=True).replace(" ","").replace("\n", "").replace("\t", ""))
            excel_sheet.append(data2)
        for cell in excel_sheet[row] :
            cell.font = normal_font
            cell.alignment = wrap_alignment
        row += 1
excel_file.save("mysite.xlsx")
excel_file.close()
print(f"mysite.xlsx 저장완료")
```
